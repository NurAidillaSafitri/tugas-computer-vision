```python
b, g, r = cv2.split(img)
print('dimensi gambar:')
print(img.shape)
```

```python
img3 = img
for i, row in enumerate(img):
  for j, pixel in enumerate(img):
    means = (b[i][j]+g[i][j]+r[i][j])/5
    img3[i][j] = [means, means, means]

cv2_imshow(img3)
```
